```
data['Rating']=np.log1p(data['Rating'])
```

```
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.boxplot(x='Rating',data=data,color='lightgreen')
plt.title('boxplot of App Ratings')
plt.xlabel('Rating')
plt.show()
```

```
from numpy._core.defchararray import lower
Q1=data['Rating'].quantile(0.25)
Q3=data['Rating'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
```

```
filterdata=data[(data['Rating']>=lower_bound) & (data['Rating']<=upper_bound)]
```

```
plt.figure(figsize=(10,6))
sns.boxplot(x='Rating',data=filterdata,color='lightgreen')
plt.title('boxplot of App Ratings')
plt.xlabel('Rating')
plt.show()
```

```
plt.figure(figsize=(10,6))
sns.countplot(y='Category',data=data,palette='viridis')
plt.title('Category Distribution')
plt.xlabel('Count')
plt.ylabel('Category')
```

```
plt.figure(figsize=(10,6))
sns.countplot(y='Category',data=data,order=data['Category'].value_counts().index,palette='pastel')
plt.title('Category Distribution')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/Google_data.csv')

# ----- CLEANING -----

# Remove duplicate rows
df = df.drop_duplicates()

# Remove rows with missing Category values
df = df.dropna(subset=['Category'])

# Strip spaces from Category names (if any)
df['Category'] = df['Category'].str.strip()

# ----- FILTERING -----

# Take top 10 most common categories
top10 = df['Category'].value_counts().head(10)

# ----- PIE CHART -----

plt.figure(figsize=(8, 8))
plt.pie(top10, labels=top10.index, autopct='%1.1f%%')
plt.title("Top 10 App Categories (Cleaned Data)")
plt.show()
```
